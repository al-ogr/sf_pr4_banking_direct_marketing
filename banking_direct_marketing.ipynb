## Первичная обработка данных

В первичной обработки данных производится заполнение пропусков и очистка от выбросов.

Признак balance, содержащий информацию о балансе клиента, имеет тип данных object, в качестве разделителя целой и дробной частей числа использована запятая, для разделенния тысячных числа использован пробел, в конце значения через пробел указана валюта (знак **$**). Признак в своей сути является непрерывным числовым и требует соответствующих преобразований.

```python
# Преобразование balance во float
df_clients['balance'] = df_clients['balance'].apply(lambda x: x.replace('$', '').replace(',', '.').replace(' ', '') \
                                                              if not pd.isna(x) else x)
df_clients['balance'] = df_clients['balance'].astype(float)
```

```python
df_clients['balance'].describe()
```

Ввиду небольшого количества пропусков признака balance (25 из 11162) целесообразно их заполнить медианным значением.

```python
# Обработка пропусков balance
median_balance = df_clients['balance'].median()
df_clients['balance'].fillna(median_balance, inplace=True)
df_clients['balance'].describe()
```

После заполнения пропусков признака balance его статистические данные существенно не изменились.

В соответствии со справочной информацией клиенты подразделяются на следующие возрастные категории:
- менее 22 лет (молодежь)
- 22-34 лет (молодые люди, образовавшие семью)
- 35-54 лет (лица "зрелого возраста")
- 55 и более лет (лица, готовящиеся к уходу на пенсию)

Распределение на категории по данным признакам вносится в новые: **CreditScoreCat** и **AgeCat** соответственно.

```python
# создайте возрастные группы и определите, в каких группах более склонны открывать депозит, чем отказываться от предложения
# Функция распределения возраста по категориям
def get_age_cat(age):
    if age >= 30 and age < 40:
        return "30-40"
    elif age >= 40 and age < 50:
        return "40-50"
    elif age >= 50 and age < 60:
        return "50-60"
    elif age >= 60:
        return "60+"
    elif age < 30:
        return "<30" 


# Создание признака категорий возраста
df_clients['age_cat'] = df_clients['age'].apply(get_age_cat)
```
